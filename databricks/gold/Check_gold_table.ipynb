```python
%sql
-- Comparison of Raw (Bronze) vs. Aggregated (Gold) tables
SELECT 'Bronze Raw Events' as table_name, count(*) as row_count 
FROM databricks_clickstream_dev.clickstream_bronze.bronze_clickstream

UNION ALL

SELECT 'Gold Daily Active Users' as table_name, count(*) as row_count 
FROM databricks_clickstream_dev.clickstream_gold.daily_active_users

UNION ALL

SELECT 'Gold Traffic Insights' as table_name, count(*) as row_count 
FROM databricks_clickstream_dev.clickstream_gold.traffic_performance;
```

```python
%sql
-- Logic & Insight Check. This query tests if your "Slicing and Dicing" capability works. It answers a real business question: "Which country is driving the most traffic today, and what is their engagement score?"
SELECT 
    country, 
    sum(unique_users) as total_visitors, 
    round(avg(engagement_score), 2) as avg_clicks_per_user
FROM databricks_clickstream_dev.clickstream_gold.traffic_performance
WHERE day = 27 -- Or your most recent data day
GROUP BY country
ORDER BY total_visitors DESC;
```
